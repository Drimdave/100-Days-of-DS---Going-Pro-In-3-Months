# Show the number of confirmed, deaths and recovered in each Region

```python
df.groupby('Region').sum().head(20)
```

# Showing the top 20 Regions with high confirmed cases

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending = False).head(20)
```

# Showing the top 20 Regions with low confirmed cases

```python
df.groupby('Region')['Confirmed'].sum().sort_values(ascending = True).head(20)
```

# Showing the top 20 Regions with high death rates

```python
df.groupby('Region')['Deaths'].sum().sort_values(ascending =False).head(20)
```

# Showing the top 20 Regions with low death rates

```python
df.groupby('Region')['Deaths'].sum().sort_values(ascending = True).head(20)
```

# Showing the top 20 Regions with high recovery rates

```python
df.groupby('Region')['Recovered'].sum().sort_values(ascending =False).head(20)
```

# Showing the top 20 Regions with low recovery rates

```python
df.groupby('Region')['Recovered'].sum().sort_values(ascending = True).head(20)
```

# Showing all the columns together

```python
df.groupby('Region')['Confirmed', 'Deaths', 'Recovered'].sum().head(20)
```
