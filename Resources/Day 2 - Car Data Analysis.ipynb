# Fill all the Null Values

```python
dt.columns
```

```python
dt['Make'] = dt['Make'].fillna(method = 'bfill')
dt['Model'] = dt['Model'].fillna(method = 'ffill')
dt['Type'] = dt['Type'].fillna(method = 'bfill')
dt['Origin'] = dt['Origin'].fillna(method = 'bfill')
dt['DriveTrain'] = dt['DriveTrain'].fillna(method = 'ffill')
dt['MSRP'] = dt['MSRP'].fillna(method = 'bfill')
dt['Invoice'] = dt['Invoice'].fillna(method = 'bfill')
dt['EngineSize'] = dt['EngineSize'].fillna(method = 'bfill')
dt['Cylinders'] = dt['Cylinders'].fillna(method = 'ffill')
dt['Horsepower'] = dt['Horsepower'].fillna(method = 'bfill')
dt['MPG_City'] = dt['MPG_City'].fillna(method = 'ffill')
dt['MPG_Highway'] = dt['MPG_Highway'].fillna(method = 'bfill')
dt['Weight'] = dt['Weight'].fillna(method = 'ffill')
dt['Wheelbase'] = dt['Wheelbase'].fillna(method = 'bfill')
dt['Length'] = dt['Length'].fillna(method = 'ffill')
```
